##### Introduction to Functions
Definition:

A function is a block of code that performs a specific task.
Functions help in organizing code, reusing code, and improving readability.

```python
# Syntax

def function_name(parameters):
    """Docstring"""
    # Function body
    return expression
```

```python
## Why functions?

num = 24
if (num % 2 == 0):
    print("The number is even")
else:
    print("The number is odd")
```

```python
def even_or_odd(num):
    """This function finds even or odd"""
    if (num % 2 == 0):
        print("The number is even")
    else:
        print("The number is odd")
        
```

```python
## Call the function

even_or_odd(24)
```

```python
even_or_odd(23)
```

```python
## function with multiple parameters

def add(a,b):
    c = a + b
    return c

add(3,5)
```

```python
## Default parameters

def greet(name):
    print(f"Hello {name}")

greet("Tamas")
```

```python
def greet(name):
    print(f"Hello {name}")

greet()
```

```python
def greet(name="Guest"):
    print(f"Hello {name}")

greet()
```

```python
def greet(name="Guest"):
    print(f"Hello {name}")

greet("Tamas")
```

```python
## Variable Length Arguments
# Positional and Keyword arguments

def print_numbers(*args):
    for number in args:
        print(number)

print_numbers(1,2,3,4,5,6,7,8,"Tamas")
```

```python
### Keywords arguments

def print_details(**kwargs):
    for key,value in kwargs.items():
        print(f"{key}:{value}")

print_details(name="Tamas", age="32", country="Slovakia")
```

```python

def print_details(*args,**kwargs):
    for val in args:
        print(f"Positional argument: {val}")

    for key,value in kwargs.items():
        print(f"{key}:{value}")

print_details(1,2,3,4,name="Tamas", age="32", country="Slovakia")
```

```python
## Return Statements

def multiply_func(a,b):
    return a*b

multiply_func(2,3)
```

```python
## Return Multiple Parameters

def multiply_func(a,b):
    return a*b, a

multiply_func(2,3)
```
