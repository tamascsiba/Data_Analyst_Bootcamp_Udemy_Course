##### Introduction to Tuples
Explanation:

Tuples are ordered collections of items that are immutable.
They are similar to lists, but their immutability makes them different.


```python
## Create a tuple


empty_tuple=()
print(empty_tuple)
print(type(empty_tuple))
```

```python
lst=list()
print(type(lst))
print(lst)
```

```python
numbers=tuple([1,2,3,4,5,6])
numbers
```

```python
list((1,2,3,4,5,6))
```

```python
mixed_tuple = (1,"Hello World", 3.14, True)
print(mixed_tuple)
```

```python
## Accessing Tuple Elements

numbers
```

```python
print(numbers[0])
print(numbers[-1])
```

```python
numbers[0:4]
```

```python
numbers[::-1]
```

```python
## Tuple Operations

concatenation_tuple = mixed_tuple + numbers
print(concatenation_tuple)
```

```python
mixed_tuple * 3
```

```python
numbers * 3
```

```python
## Immutable Nature of Tuples
## Tuples are immutable, meaning their elements cannot be changed once assigned

lst = [1,2,3,4,5]
print(lst)

lst[1] = 'Tamas'
print(lst)
```

```python
numbers[1] = 'Tamas'
```

```python
numbers
```

```python
## Tuple methods

print(numbers.count(1))
print(numbers.index(3))
```

```python
## Packing and Unpacking Tuple

## Packing
packed_tuple = 1,"Hello",3.14,True
print(packed_tuple)
```

```python
## Unpacking

a,b,c,d = packed_tuple
print(a)
print(b)
print(c)
print(d)
```

```python
## Unpacking with *

numbers = (1,2,3,4,5,6)
first,*middle,last = numbers
print(first)
print(middle)
print(last)
```

```python
## Nested Tuple

nested_list = [[1,2,3,4],[5,6,7,8],[1,"Hello",3.14,True]]

nested_list[0][0:3]
```

```python
nested_list_with_tuple = [[1,2,3,4],[5,6,7,8],(1,"Hello",3.14,True)]

nested_list_with_tuple[2][0:3]
```

```python
nested_tuple = [(1,2,3,4),(5,6,7,8),(1,"Hello",3.14,True)]
print(nested_tuple[0])
print(nested_tuple[1][2])
```

```python
## Iterationg over nested tuples

for sub_tuple in nested_tuple:
    print("\n")
    for item in sub_tuple:
        print(item, end= " ")
```
